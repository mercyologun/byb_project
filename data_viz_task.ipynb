```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('Cars93.csv')
```

```python
df.columns
```

### 1. Boxplot

```python
# Boxplot for revs per mile
sns.boxplot(data=df['Rev.per.mile'])
plt.title('Boxplot of Revs per Mile')
plt.xlabel('Revs per Mile')
plt.show()
```

```python
sns.boxplot(data=df, y='Rev.per.mile', x='Manufacturer')
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?

### 2. Histogram

```python
# Histogram of MPG
sns.histplot(data=df['MPG.city'], bins=20, kde=True)
plt.title('Histogram of MPG city')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data=df['MPG.highway'], bins=20, kde=True)
plt.title('Histogram of MPG highway')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer: on the highway**

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(x='Wheelbase', y='Turn.circle', data=df)
plt.title('Wheelbase vs Turning Circle')
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer: the turning circle gets bigger**  


### 4. Barplot

```python
# Barplot for mean horsepower
plt.bar(['Horsepower'], df['Horsepower'].mean())
plt.title('Mean Horsepower')
plt.ylabel('Horsepower')
plt.show()
```

```python
sns.barplot(data=df, y='Horsepower', x='Type')
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer: yes**
